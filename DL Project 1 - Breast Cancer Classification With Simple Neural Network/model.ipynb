Breast Cancer Classification with Simple Neural Network
=======================================================

Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

Data Collection and Processing

```python
# Load the dataset
breast_cancer = datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
#loading the dataset to a pandas DataFrame
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
#print the first 5 rows of the data
data.head()
```

```python
# Adding the target column to the DataFrame
data['target'] = breast_cancer.target
```

```python
#print the first 5 rows of the data
data.head()
```

```python
#shape of the data
data.shape
```

```python
#getting some information about the data
data.info()
```

```python
#checking missing values
data.isnull().sum()
```

```python
#statistical measures of the data
data.describe()
```

```python
#checking the distribution of target variable
data['target'].value_counts()
```

1. Benign -->1
2. Malignant -->0

```python
data.groupby('target').mean()
```

Separating Features and Target

```python
X = data.drop('target', axis=1)
Y = data['target']
```

```python
print(X)
print(Y)
```

Splitting the Data into Training and Testing Sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#print the shape of X_train and X_test
print(X.shape, X_train.shape, X_test.shape)
```

Standardizing the Data

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaler.transform(X_train)
scaler.transform(X_test)
```

Building the Neural Network Model

```python
tf.random.set_seed(3)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(X_train.shape[1],)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
#compiling the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

Train the Model

```python
record = model.fit(X_train, Y_train,validation_split=0.1, epochs=10)
```

```python
plt.plot(record.history['accuracy'])
plt.plot(record.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

Training the Neural Network Model

```python
plt.plot(record.history['loss'])
plt.plot(record.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')

plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

Accuracy of the Model

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print('Test accuracy: ', accuracy)
```

Building the prediction System

```python
#Sample data
sample_data = X_test.iloc[0]
```

```python
input_data = (sample_data.values).reshape(1, -1)

#change the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data).reshape(1, -1)

#standardize the input data
std_data = scaler.transform(input_data_as_numpy_array)

prediction = model.predict(std_data)

prediction_label = [np.argmax(prediction[0])]
print(prediction_label)

if prediction_label[0] == 0:
    print('The cancer is malignant')
else:
    print('The cancer is benign')
    
```
